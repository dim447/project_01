Загружаем данные Фитофарм о продажах на Озон за 3 месяца лета

```python
import pandas as pd

df= pd.read_csv('/content/sample_data/orders.csv', sep=';')
df.head(5)
```

Смотрим названия колонок и удаляем не нужные колонки

```python
df.columns
```

Сохраняем новый DataFrame как df1

```python
df1 = df.drop(['Номер заказа', 'Номер отправления', 'Принят в обработку',
       'Дата отгрузки', 'Статус', 'Дата доставки', 'Сумма отправления',
       'Код валюты отправления','OZON id', 'Артикул','Связанные отправления', 'Выкуп товара',
       'Цена товара до скидок', 'Скидка %', 'Скидка руб', 'Акции',
       'Объемный вес товаров, кг','Способ доставки', 'Сегмент клиента', 'Юридическое лицо',
       'Способ оплаты', 'Штрихкод ювелирного изделия','Код валюты товара','Стоимость доставки'], axis=1)
df1
```

Смотрим общую статистику числовых значений: "Количество" и "Итоговая стоимость товара"

```python
df1.describe()
```

Можно посчитать общую сумму продаж .sum()

```python
df1['Итоговая стоимость товара'].sum()
```

Проверяем склады отгрузки, получаем массив

```python
df1['Склад отгрузки'].unique()
```

Получаем количество отгруженных товаров по складам отгрузки и сортируем по убыванию

```python
sklad_otgruzki = df1.groupby('Склад отгрузки')['Количество'].count()
sklad_otgruzki
```

Самый загруженный склад - Жуковский, а регион Московский

```python
sklad_otgruzki.sort_values(ascending=False)
```

Группируем по Региону доставки и считаем количество товара

```python
df1.groupby(['Регион доставки'])['Количество'].count().sort_values(ascending=False)
```

Проверяем названия товара на уникальность

```python
df1['Наименование товара'].unique()
```

Выводим список товаров по убыванию: самый популярный чай - "Друг Вашего желудка"

```python
df1.groupby('Наименование товара')['Количество'].count().sort_values(ascending=False)
```

Также выводим статистические данные о стоимости товара: средняя, медианная, минимальная и максимальная..

```python
df1.groupby('Наименование товара')['Итоговая стоимость товара'].describe()
```

Также можно вывести стат данные по каждой позиции.

```python
df1[df1['Наименование товара'] == 'Чай травяной Друг Вашего иммунитета ф/п 1.5г №20 /20 пакетиков Фитосбор, фиточай, БАД'].describe()
```
